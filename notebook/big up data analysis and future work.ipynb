```python
#coding=utf-8
import os
import sys
from os.path import abspath, dirname
sys.path.insert(0, dirname(dirname(os.path.abspath('上证指数大涨5.6的后市分析.ipynb'))))
import const as ct
import pandas as pd
from cindex import CIndex
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib import ticker as mticker
from mpl_finance import candlestick2_ochl
from common import get_day_nday_after, get_dates_array, get_day_nday_ago, get_dates_array
```

```python
def plot_index(mdate, day_num = 100):
    base_color = '#e6daa6'
    name = "shanghai_%s_image" % mdate
    start_date = get_day_nday_ago(mdate, num = day_num, dformat = "%Y-%m-%d")
    end_date = get_day_nday_after(mdate, num = day_num, dformat = "%Y-%m-%d")
    date_arrary = get_dates_array(start_date, end_date)
    date_index = 0
    i_data = index_obj.get_k_data_in_range(start_date, end_date)
    i_data = i_data[['date', 'open', 'high', 'low', 'close', 'volume', 'amount']]
    fig, index_ax = plt.subplots(figsize = (16, 10))
    fig.subplots_adjust(bottom = 0.2)
    width = 5
    height = i_data.loc[i_data.date == mdate].high.values[0] - i_data.loc[i_data.date == mdate].low.values[0]
    x = i_data.loc[i_data.date == mdate].index.values[0] - width / 2 
    y = i_data.loc[i_data.date == mdate].low.values[0]
    candlestick2_ochl(index_ax, i_data['open'], i_data['close'], i_data['high'], i_data['low'], width = 1.0, colorup = 'r', colordown = 'g')
    index_ax.add_patch(patches.Rectangle(
                        (x, y),   # (x,y)
                        width,    # width
                        height,   # height
                        linewidth = 2,
                        edgecolor = 'b',
                        facecolor = 'none'))
    index_ax.set_ylabel(name)
    index_ax.set_xticks(range(0, len(i_data['date']), 10))
    fig.autofmt_xdate()
    plt.show()
```

```python
index_obj = CIndex('000001', dbinfo = ct.OUT_DB_INFO, redis_host = '127.0.0.1')
df = index_obj.get_k_data()
cdates = df[(df.pchange > 4) & (df.date > '1999-01-01')].date.tolist()
print(cdates)
```

结论：总结过去20年以来的所有数据：
1、大盘超过4%的单日大涨，基本没有任何长期意义。
2、如果是牛市（1999,2007和2015年的大牛市，貌似8年一次大牛市），会连续出现超过4%以上的大涨，至少3次以上。并且大涨之后，不回调。
3、对于熊市后期，出现第一次大盘大涨时，是一个好的高抛低吸的机会。

所以，对于我们，需要观察两个指标：
1、大涨之后，大盘是否会回调，跌破大涨的价格。
2、是否在短期内出现多次大涨。
3、目前距离上一次牛市才只有4年，数据上还不支持。
